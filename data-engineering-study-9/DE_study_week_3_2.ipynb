### 사용자별로 처음과 마지막 채널 알아내기
* 사용자ID, 첫번째 채널, 마지막 채널
* ROW_NUMBER를 이용해서 해보자

```python
%%sql
SELECT ts, channel
FROM raw_data.user_session_channel usc
JOIN raw_data.session_timestamp st ON usc.sessionid = st.sessionid
WHERE userid = 251
ORDER BY 1;
```

```python
# ROW_NUMBER로 seq 열 추가하기
%%sql
SELECT ts, channel, ROW_NUMBER () OVER (PARTITION BY usc.userid ORDER BY ts ASC) seq
FROM raw_data.user_session_channel usc
JOIN raw_data.session_timestamp st ON usc.sessionid = st.sessionid
WHERE userid = 251
ORDER BY 1;
```

```python
# 첫번째 채널 ROW_NUMBER, 마지막 채널 ROW_NUMBER를 한 테이블로 만들기

%%sql

CREATE TABLE soooojin_lee.channel_seq AS 
SELECT 
  ts, 
  userid,
  channel, 
  ROW_NUMBER () OVER (PARTITION BY usc.userid ORDER BY ts ASC) first_seq,
  ROW_NUMBER () OVER (PARTITION BY usc.userid ORDER BY ts DESC) last_seq
FROM raw_data.user_session_channel usc
JOIN raw_data.session_timestamp st ON usc.sessionid = st.sessionid
ORDER BY 1;
```

```python
%%sql

SELECT * FROM soooojin_lee.channel_seq LIMIT 10;
```

```python
# self join 결과 확인

%%sql

SELECT *
FROM soooojin_lee.channel_seq first_channel
JOIN (SELECT userid, channel FROM soooojin_lee.channel_seq WHERE last_seq = 1) last_channel
ON first_channel.userid = last_channel.userid
WHERE first_channel.first_seq = 1
LIMIT 10;
```

```python
## 최종 결과 쿼리
%%sql

SELECT first_channel.userid, first_channel.channel AS first, last_channel.channel AS last
FROM soooojin_lee.channel_seq first_channel
JOIN (SELECT userid, channel FROM soooojin_lee.channel_seq WHERE last_seq = 1) last_channel
ON first_channel.userid = last_channel.userid
WHERE first_channel.first_seq = 1;
```
